**1. Import libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**2. Bar Chart (Comparing values)**

```python
products = ['Laptop', 'Mobile', 'Tablet', 'Headphones']
sales = [120, 200, 150, 90]

plt.bar(products, sales)
plt.title('Product Sales Comparison')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

**3. Line Chart**

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
revenue = [2000, 2500, 2200, 2700, 3000]

plt.plot(months, revenue, marker='o')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
```

**4. Pie Chart(Composition)**

```python
labels = ['Online', 'Store', 'Partners']
sales = [45, 30, 25]

plt.pie(sales, labels=labels, autopct='%1.1f%%')
plt.title('Sales Distribution')
plt.show()
```

**5. Histogram**

```python
data = np.random.normal(50,10,200)

plt.hist(data, bins=20)
plt.title('Distribution of Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

**6. Scatter Plot (Realtionship)**

```python
hours = [1,2,3,4,5,6,7]
marks = [40, 45, 50, 55, 65, 70, 80]

plt.scatter(hours, marks)
plt.title('Study Hours vs Marks')
plt.xlabel('Hours studied')
plt.ylabel('Marks')
plt.show()
```

**7. Seaborn Visualization**

```python
iris = sns.load_dataset('iris')
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
plt.title('Seaborn Scatter Plot - Iris Dataset')
plt.show()
```
